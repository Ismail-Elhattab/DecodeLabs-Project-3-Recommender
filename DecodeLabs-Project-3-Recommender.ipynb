```python
# ==========================================
# DECODELABS PROJECT 3: TECH STACK RECOMMENDER
# ==========================================

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# --- SIMULATED DATASET ---
# Simulating the 'raw_skills.csv' mentioned in the project manual.
# In a real scenario, you would use: df = pd.read_csv('raw_skills.csv')
data = {
    'Job_Role': [
        'DevOps Engineer',
        'Cloud Architect',
        'Backend Developer',
        'Data Scientist',
        'Frontend Developer',
        'Sys Admin'
    ],
    'Skills': [
        'Python Cloud Automation AWS Docker Kubernetes CI/CD',
        'Cloud AWS Azure Infrastructure Networking Security',
        'Java Python SQL APIs Database Server',
        'Python SQL Machine Learning Data Analysis TensorFlow Pandas',
        'JavaScript HTML CSS React UI UX Web Design',
        'Linux Networking Automation Bash Security Server'
    ]
}
df = pd.DataFrame(data)

print("--- DIGITAL MATCHMAKER ENGINE INITIALIZED ---\n")

# ==========================================
# STEP 1: INGESTION (User State)
# ==========================================
# The project requires accepting a minimum of 3 user inputs.
user_skills_input = ["Python", "Cloud", "Automation"]
user_profile_string = " ".join(user_skills_input)

print(f"1. INGESTION: Captured user skills -> {user_skills_input}")

# ==========================================
# STEP 2: SCORING (Similarity Logic)
# ==========================================
# Upgrading feature extraction with TF-IDF Weighting.
# This creates our numerical array (vectors) within a shared vocabulary space.
vectorizer = TfidfVectorizer()

# We fit the vectorizer on the job skills, and append the user's profile to the end
# so they share the exact same vocabulary space.
all_text_data = df['Skills'].tolist() + [user_profile_string]
tfidf_matrix = vectorizer.fit_transform(all_text_data)

# Separate the item vectors (job roles) and the user vector
job_vectors = tfidf_matrix[:-1]
user_vector = tfidf_matrix[-1]

# Calculate Cosine Similarity (measuring the mathematical angle between vectors)
# This will return a score between 0 (Orthogonal) and 1 (Aligned).
similarity_scores = cosine_similarity(user_vector, job_vectors).flatten()

print("2. SCORING: TF-IDF applied and Cosine Similarity calculated.")

# ==========================================
# STEP 3: SORTING
# ==========================================
# Add the scores back to our dataframe so we can rank them.
df['Similarity_Score'] = similarity_scores

# Organize the scored dataset in descending order.
df_sorted = df.sort_values(by='Similarity_Score', ascending=False)

print("3. SORTING: Data structures ranked in descending order.")

# ==========================================
# STEP 4: FILTERING (Top-N List)
# ==========================================
# Prevent choice overload by truncating the output to the Top 3 items.
TOP_N = 3
top_recommendations = df_sorted.head(TOP_N)

print("4. FILTERING: Truncating to Top 3 to prevent choice overload.\n")

# --- FINAL OUTPUT ---
print("==========================================")
print("🎯 TOP RECOMMENDED CAREER PATHS")
print("==========================================")
for index, row in top_recommendations.iterrows():
    # We round to 4 decimal places for clean viewing
    print(f"Role: {row['Job_Role']} | Match Score: {row['Similarity_Score']:.4f}")
```
